````python
import pandas as pd
import pathlib
from collections import Counter

CURRENT_PATH = pathlib.Path.cwd()
DATA_PATH = 'data/data.csv'
FULL_DATA_PATH = CURRENT_PATH.joinpath(DATA_PATH)


def load_csv_data(full_data_path=FULL_DATA_PATH):
    return pd.read_csv(full_data_path, index_col='Responder_id')
````

````python
df = load_csv_data()
df.head()
````

````python
languages_counter = Counter()

df['LanguagesWorkedWith'].apply(lambda x: languages_counter.update(x.split(';')))

most_common_lan = languages_counter.most_common(5)
most_common_lan
````

````python
labels = []
slices = []

for lab, sli in most_common_lan:
    labels.append(lab)
    slices.append(sli)

    
plt.style.use('fivethirtyeight')

plt.pie(slices, labels=labels, wedgeprops={'edgecolor':'black'})


plt.title('My Awesome Pie Chart')
plt.tight_layout()
plt.show()
````

# Emphasis a slice.

A lot of times you'll see some pie charts that have an emphasis on one piece of the chart.

Let's say that we were preparing a talk for a Python conference and we wanted to emphasize the Python slice for the chart so to do this we can pass an ```python explode``` argument and ```python This will be a list of values``` that will offset the slice.

````python
labels = []
slices = []

# Explode list. These values will be a list of floats which represent how much we want to emphasize that slice.
# So a zero 0 will keep that slice how it is and any number besides zero will represent the fraction of the radius that we'd like to explode the value.
# 0.1 is 10% of the radius exploded out
explode = [0, 0, 0, 0.1, 0]

for lab, sli in most_common_lan:
    labels.append(lab)
    slices.append(sli)

    
plt.style.use('fivethirtyeight')

plt.pie(slices, labels=labels, explode=explode, wedgeprops={'edgecolor':'black'})


plt.title('My Awesome Pie Chart')
plt.tight_layout()
plt.show()
````
