## Format our plot to make this look a bit better.

### .autofmt_xdate()

This will rotate our dates so that they fit a bit nicer and change their alignments and things like that.

Now we haven't talked much about figures and axis in this series yet, that is gonna be in the sub plots video in a couple more videos but basically this is going to be a method on our figure and not on this pyplot object that we have been using (plt.) so to get the current figure from pyplot we can say:

````python
plt.style.use('seaborn')

dates = [
    datetime(2019, 5, 24),
    datetime(2019, 5, 25),
    datetime(2019, 5, 26),
    datetime(2019, 5, 27),
    datetime(2019, 5, 28),
    datetime(2019, 5, 29),
    datetime(2019, 5, 30),
]

y = [0, 1, 3, 4, 6, 5, 7]

plt.plot_date(dates, y, linestyle='solid')



# .gcf() = get current figure 
plt.gcf().autofmt_xdate()



# plt.legend()
# plt.title('fff')
# plt.xlabel('ffff')
# plt.ylabel('llll')

plt.tight_layout()
plt.show()
````

Then we can see that now these dates are rotated and they have different alignments. That just makes it so that these dates are not so bunched together and it makes it easier to read.

## Change the format of our dates.

So what if instead of how they're displayed now with the year-month-day what if instead we wanted them to start with the name of the month and then the day and then the year so to do this we have to use some datetime formatting.

We will use ```from matplotlib import dates as mpl_dates``` and from that imported module we're going to use the date formatter class and we're going to be passing in any format string that you could also pass into the strf time method from the datetime class.

````python
plt.style.use('seaborn')

dates = [
    datetime(2019, 5, 24),
    datetime(2019, 5, 25),
    datetime(2019, 5, 26),
    datetime(2019, 5, 27),
    datetime(2019, 5, 28),
    datetime(2019, 5, 29),
    datetime(2019, 5, 30),
]

y = [0, 1, 3, 4, 6, 5, 7]

plt.plot_date(dates, y, linestyle='solid')



# .gcf() = get current figure 
plt.gcf().autofmt_xdate()



date_format = mpl_dates.DateFormatter('%b, %d %Y')
# Now we need to set this as the format for our x-axis
# Grab the axis to run the format method 
# .gca() = get current axis
# format the x-axis by saying .xaxis.set_major_formatter(date_format)
plt.gca().xaxis.set_major_formatter(date_format)


# plt.legend()
# plt.title('fff')
# plt.xlabel('ffff')
# plt.ylabel('llll')

plt.tight_layout()
plt.show()
````

Then we can see that we don't have that Year-Month-Day that we had before, now this is formatted to say Month, Day Year.

# Real-world example.

````python
import pathlib
````

````python
CWD_PATH = pathlib.Path.cwd()
DATA_PATH = 'data/data3.csv'
FULL_DATA_PATH = CWD_PATH.joinpath(DATA_PATH)


def load_data(full_data_path=FULL_DATA_PATH):
    d_parser = lambda x: datetime.strptime(x, '%Y-%m-%d')
    return pd.read_csv(full_data_path, parse_dates=['Date'], date_parser=d_parser)


df = load_data()
df.head()
````

````python
plt.style.use('seaborn')

df.sort_values(by='Date', inplace=True)

price_date = df['Date']
price_close = df['Close']
price_open = df['Open']

plt.plot_date(price_date, price_close, linestyle='solid')
plt.plot_date(price_date, price_open, linestyle='solid')

# plt.legend()
# plt.title('fff')
# plt.xlabel('ffff')
# plt.ylabel('llll')

plt.tight_layout()
plt.show()
````

````python
plt.style.use('seaborn')

df1 = pd.read_csv('data/data3.csv')

df1['Date'] = pd.to_datetime(df1['Date'])
df1.sort_values(by='Date', inplace=True)

price_date = df1['Date']
price_close = df1['Close']

plt.plot_date(price_date, price_close, linestyle='solid')

plt.gcf().autofmt_xdate()

format_date = mpl_dates.DateFormatter('%b, %d %Y')
plt.gca().xaxis.set_major_formatter(format_date)


# plt.legend()
plt.title('BTC Prices')
plt.xlabel('Date')
plt.ylabel('Closing Price')

plt.tight_layout()
plt.show()
````
