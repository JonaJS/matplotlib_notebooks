# Scatter Plots

Scatter plots are great when you want to show the relationship between two sets of values and see if they are correlated.

````python
import pathlib
import pandas as pd
from matplotlib import pyplot as plt
````

````python
plt.style.use('seaborn')

x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]

plt.scatter(x, y)

plt.tight_layout()
plt.show()
````

We can see that we have a scatter plot of these random values, now scatter plots are really nice for seeing different trends or outliers or things like that now since these are random then we don't really have any trends here but sometimes that's important to know also that there's not a trend.

So our scatter plot since it looks random like this then this likely tells us that there's no correlation between our two lists of random values and we'll see dataset here in a bit that is more correlated with the values but now let's look at some basic customizations that we can make to our scatter plot.

## Size of the dots.

````python
plt.style.use('seaborn')

x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]

plt.scatter(x, y, s=100)

plt.tight_layout()
plt.show()
````

## Change the color and marker styles.

````python
plt.style.use('seaborn')

x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]

plt.scatter(x, y, s=100, c='green', marker='x')

plt.tight_layout()
plt.show()
````

## Adding edges to the circles and some alpha

````python
plt.style.use('seaborn')

x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]

plt.scatter(x, y, s=100, c='green', edgecolor='black', linewidth=1, alpha=0.75)

plt.tight_layout()
plt.show()
````

## How the colors and sizes can actually be on a per mark basis rather than applying them to all of the marks.

What would you want multiple colors or sizes?

R= Well having the ability to have multiple colors and sizes actually allows us to add additional datasets into our plot, so for example let's say that we have our current plot that we just looked at but we wanted to add some additional information.

For example, let's pretend that our current plot is some survey data about a bunch of people and we wanted to break down the data further into something more specific.

So for example, let's say that we had these people rate something from 1 to 10 and we wanted to somehow plot their rating as well. Well to do that we could simply assign different numbers to these different possibilities and those will then give you different colors on your scatter plot as long as you pass that into your method:

````python
plt.style.use('seaborn')

x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]

colors = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]

plt.scatter(x, y, s=100, c=colors, edgecolor='black', linewidth=1, alpha=0.75)

plt.tight_layout()
plt.show()
````

So what this is doing here is whenever we plot this X and Y value here of 5 and 7 it also has a color of 7 and these different colors here 7, 5, 9 all of those are going to correspond to different colors here on our chart.

Now I really don't like the colors that we're getting here these are just shades of gray, we can actually change these by using a color map and just like the marker symbols there are a ton of built-in color maps that we can use.

````python
plt.style.use('seaborn')

x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]

colors = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]

plt.scatter(x, y, s=100, c=colors, cmap='Greens', edgecolor='black', linewidth=1, alpha=0.75)

plt.tight_layout()
plt.show()
````

Now we can see that now we're getting these different shades of green as the intensity so I think the lighter ones are closer to 0 and the dark ones are closer to 10 but we really don't know based on how this is represented here so you're probably going to want to add a label for your color map too so that people viewing our chart know what these colors represent.

## Add a label.

````python
plt.style.use('seaborn')

x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]

colors = [9, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]

plt.scatter(x, y, s=100, c=colors, cmap='Greens', edgecolor='black', linewidth=1, alpha=0.75)
cbar = plt.colorbar()
cbar.set_label('Satisfaction')

plt.tight_layout()
plt.show()
````

Now we can see that we have a color bar here on the right side and now we have multiple points of information here so we can se our XY data but then we can also see depending on how these are colored is that person's satisfaction level to these dark ones are very satisfied and the lighter color ones were not satisfied.


Now we can also change the sizes of our data points as well so just like with the color, this can add in another way of explaining our data even further. For example, I see a lot of scatter plots that use the size of the dot for thins like population or things like that or maybe even the sample size for that data point.

So I've got a random list of sizes commented out down here as well and just like with the colors this is a list of 20 different sizes here that correspond to the data points and the x and y variables that were plotting so if we were to pass in these sizes list here as my sizes:

````python
plt.style.use('seaborn')

x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]

colors = [9, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]
sizes = [209, 486, 381, 255, 191, 315, 185, 228, 174, 538, 239, 394, 399, 153, 273, 293, 436, 501, 397, 539]

plt.scatter(x, y, s=sizes, c=colors, cmap='Greens', edgecolor='black', linewidth=1, alpha=0.75)
cbar = plt.colorbar()
cbar.set_label('Satisfaction')

plt.tight_layout()
plt.show()
````

We can see that now each of these individual plots have a size as well and we can use that for different types of data.

## Plotting some real-world data from a CSV.

````python
CWD_PATH = pathlib.Path.cwd()
DATA_PATH = 'data/2019-05-31-data.csv'
FULL_DATA_PATH = CWD_PATH.joinpath(DATA_PATH)


def load_data(full_data_path=FULL_DATA_PATH):
    return pd.read_csv(full_data_path)


df = load_data()
df.head()
````

````python
plt.style.use('seaborn')

view_count = df['view_count']
likes = df['likes']
ratio = df['ratio']

plt.scatter(view_count, likes, edgecolor='black', linewidth=1, alpha=0.75)

plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.tight_layout()
plt.show()
````

We can see here that we get a scatter plot, now like I said this is the top 200 videos on the trending page so there should be 200 dots here. Now it looks like some of our data is bunched here in the bottom left and that's because we have one outlier here at the top right that was a video that had a lot more views and a lot more likes than the other view on the trending page (actually went back to the original data to see what video was messing up my nice little scatter plot here and it was the new old town road music video by Lil Nas X and Billy Ray Cyrus so that's who to blame for that outlier but I'm actually glad that there was an outlier because it reminded me that ```we can use a log scale with scatter plot as well to lessen how much those outliers actually skew the plot so let's make this look a bit beter``` and use a log scale for our axis we can simply say:

````python
plt.style.use('seaborn')

view_count = df['view_count']
likes = df['likes']
ratio = df['ratio']

plt.scatter(view_count, likes, edgecolor='black', linewidth=1, alpha=0.75)

# scale with log
plt.xscale('log')
plt.yscale('log')

plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.tight_layout()
plt.show()
````

Now that it's using a log scale instead of a regular scale, now we can see that those outliers don't skew the data so much and that we can kind of see the correlation better here, so the correlation between how many views a video has and how many likes it has really stands here in this plot.

So now let's also use the ratio of likes and dislikes in this plot. I think that would be a good metric to use for the color of our points and we could also try to use that for size as well but I think that the ratios might be a little bit too close for us to really tell the difference in sizes like like we can for the colors.

````python
plt.style.use('seaborn')

view_count = df['view_count']
likes = df['likes']
ratio = df['ratio']

plt.scatter(view_count, likes, c=ratio, cmap='summer', edgecolor='black', linewidth=1, alpha=0.75)

# scale with log
plt.xscale('log')
plt.yscale('log')

cbar = plt.colorbar()
cbar.set_label('Like/Dislikes Ratio')

plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.tight_layout()
plt.show()
````
